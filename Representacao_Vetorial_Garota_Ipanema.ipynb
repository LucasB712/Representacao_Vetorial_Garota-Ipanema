```python
import spacy
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nlp = spacy.load('pt_core_news_sm')
```

Pré-Processamento

```python
letra = '''Olha que coisa mais linda, mais cheia de graça
É ela, menina, que vem e que passa
Num doce balanço a caminho do mar

Moça do corpo dourado, do Sol de Ipanema
O seu balançado é mais que um poema
É a coisa mais linda que eu já vi passar

Ah, por que estou tão sozinho?
Ah, por que tudo é tão triste?
Ah, a beleza que existe
A beleza que não é só minha
Que também passa sozinha

Ah, se ela soubesse que quando ela passa
O mundo inteirinho se enche de graça
E fica mais lindo por causa do amor'''
```

```python

doc = nlp(letra)
```

```python
tokens = [token.text.lower() for token in doc if not token.is_stop and not token.is_punct]
```

```python
print("Tokens: ", tokens)
```

```python
#Vetorização usando TF- IDF (Term Frequency -- Inverse Document Frequency)
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform([' '.join(tokens)])
```

```python
# Exibindo a matriz vetorial (representação numérica do texto)
print("\nMatriz de características (TF-IDF):\n", X.toarray())
```

Palavras mais raras têm peso maior

```python

# Exibindo as palavras que correspondem a cada coluna da matriz
print("\nPalavras do vocabulário: \n", vectorizer.get_feature_names_out())
```
